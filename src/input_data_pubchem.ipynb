**This notebook prepares the original PubChem cancer drug dataset for use in machine learning. A new dataset with only the SMILES code of each molecule is created (which represents the chemical structure of the molecule), with the addition of the cancer drug status of each drug ("True").**



## PubChem
All datasets are publicly available downloaded from the PubChem visualization website or the prepared download files.

*   `pubchem_cancer_1` is from PubChem, searching for cancer, selecting compounds and exporting the dataset through the provided service

### Dataset 1 `pubchem_cancer_1`
Access this dataset [here](https://pubchem.ncbi.nlm.nih.gov/#query=cancer).

```python
pubchem_cancer_1 = pd.read_csv('/content/drive/MyDrive/Data/pubchem_cancer_1.csv',  sep = ',')
```

```python
# Add a new column with the value True
pubchem_cancer_1 = pubchem_cancer_1.assign(cancer_status = True)

# Display the modified dataset
display(pubchem_cancer_1)
```

```python
pubchem_cancer_1_inchi = pubchem_cancer_1[['cancer_status', 'inchikey']]
display(pubchem_cancer_1_inchi)
```

```python
# Define a function to convert InChIKey to SMILES
def inchikey_to_smiles(inchikey):
    try:
        compound = pcp.get_compounds(inchikey, 'inchikey')[0]
        return compound.isomeric_smiles
    except:
        return None

# Convert InChIKey to SMILES and add a new column to the dataset
pubchem_cancer_1_inchi['Smiles'] = pubchem_cancer_1_inchi['inchikey'].apply(inchikey_to_smiles)

# Display the modified dataset
display(pubchem_cancer_1_inchi)
```

```python
pubchem_cancer_1_inchi.to_csv('/content/drive/MyDrive/Data/SMILES/pubchem_cancer_1_smiles.csv', index = False)
```
